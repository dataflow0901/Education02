```
import tensorflow as tf
tf.__version__
import numpy as np
import matplotlib.pyplot as plt
import random
tf.reset_default_graph()

from tensorflow.examples.tutorials.mnist import input_data
#import tensorflow.examples.tutorials.mnist as input_data
mnist = input_data.read_data_sets('MNIST_data', one_hot=True, reshape = False)


#train_image = tf.reshape(mnist.train.images, [-1, 28*28])
#train_label = tf.reshape(mnist.train.images, [-1, 10])

x = tf.placeholder(tf.float32, shape = [None, 28*28])
y = tf.placeholder(tf.float32, shape = [None, 10])

keep_prob = tf.placeholder("float")

W1 = tf.Variable(tf.random_normal([28*28,2000], stddev=0.22), name ='weight1')
B1 = tf.Variable(tf.random_normal([2000], stddev=0.022), name ='bias1')

W2 = tf.Variable(tf.random_normal([2000,250], stddev=0.18), name ='weight2')
B2 = tf.Variable(tf.random_normal([250], stddev=0.018), name ='bias2')

W3 = tf.Variable(tf.random_normal([250,50], stddev=0.29), name ='weight3')
B3 = tf.Variable(tf.random_normal([50], stddev=0.09), name ='bias3')

W4 = tf.Variable(tf.random_normal([50,10], stddev=0.48), name ='weight4')
B4 = tf.Variable(tf.random_normal([10], stddev=0.048), name ='bias4')

#hypothesis = w*x +b
hidden1 = tf.nn.relu(tf.matmul(x,W1) + B1)
hidden1_drop = tf.nn.dropout(hidden1, keep_prob)

hidden2 = tf.nn.relu(tf.matmul(hidden1_drop,W2) + B2)
hidden2_drop = tf.nn.dropout(hidden2, keep_prob)

hidden3 = tf.nn.relu(tf.matmul(hidden2_drop,W3) + B3)

hypothesis = (tf.matmul(hidden3, W4) + B4)

cost= tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=y,logits=hypothesis))
#optimizer = tf.train.AdagradOptimizer(learning_rate=0.002)
optimizer = tf.train.AdamOptimizer(learning_rate=0.1)
train = optimizer.minimize(cost)

sess = tf.Session()

sess.run(tf.global_variables_initializer())

print("시작")
for step in range(10000):
  train_data, Y_label = mnist.train.next_batch(64)
  train_image = np.reshape(train_data, [-1, 28*28])
  train_label = np.reshape(Y_label, [-1, 10])

  cost_val, _ = sess.run([cost, train], feed_dict = {x:train_image, y:train_label, keep_prob: 0.5})

  if step %100 ==0:
    print(step, cost_val)
    
optimizer = tf.train.AdamOptimizer(learning_rate=0.001)
train = optimizer.minimize(cost) 

print("다시시작")
for step in range(10000):
  train_data, Y_label = mnist.train.next_batch(64)
  train_image = np.reshape(train_data, [-1, 28*28])
  train_label = np.reshape(Y_label, [-1, 10])
  cost_val, _ = sess.run([cost, train], feed_dict = {x:train_image, y:train_label, keep_prob: 0.5})

  if step %500 ==0:
    print(step, cost_val)


test_data, test_label = mnist.test.next_batch(1000)
test_image = np.reshape(test_data, [-1, 28*28])
test_label = np.reshape(test_label, [-1, 10])
    
correct = tf.equal(tf.argmax(hypothesis,1),tf.argmax(test_label,1))
accuracy = tf.reduce_mean(tf.cast(correct,tf.float32))

print("정확도", sess.run(accuracy, feed_dict = {x:test_image, y:test_label, keep_prob: 1.0}))
    
sess.close()
```
