```
import tensorflow as tf
tf.reset_default_graph()
```

```
X = [[60,5.5,1],[65,5.0,0],[55,6.0,1],[55,5.0,0],[70,6.5,1]]
Y = [[66,8],[74,7],[78,5],[80,4],[61,8]]

X = [[0.6,0.55,0.1],[0.65,0.5,0],[0.55,0.6,1],[0.55,0.5,0],[0.70,0.65,1]]
Y = [[0.66,0.8],[0.74,0.7],[0.78,0.5],[0.80,0.4],[0.61,0.8]]

x = tf.placeholder(tf.float32, [None, 3])
y = tf.placeholder(tf.float32, [None, 2])


w = tf.Variable(tf.random_normal([3,2],stddev=2e-3), name ='weight')
b = tf.Variable(tf.random_normal([2],stddev=2e-3),  name ='bias')

#hypothesis = w*x +b
hypothesis = tf.matmul(x,w) + b

cost = tf.reduce_mean(tf.square(hypothesis - y))

optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)

sess = tf.Session()

sess.run(tf.global_variables_initializer())

for step in range(20000):
#  cost_val, w_val, b_val, _ = sess.run([cost, w, b, train], feed_dict = {x:[1,2,3,4,5],
#                               y:[2,3,4,5,6]})
  cost_val, w_val, b_val, _ = sess.run([cost, w, b, train], feed_dict = {x:X, y:Y})

  if step %1000 ==0:
    print(step, cost_val, w_val, b_val)

#x =4,5일때 예측값
print(sess.run(hypothesis, feed_dict = {x:[[0.70, 0.6, 1],[0.655, 0.66, 0]]}))
    
sess.close()
```
